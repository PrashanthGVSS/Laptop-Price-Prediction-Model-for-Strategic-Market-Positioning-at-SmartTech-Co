```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('laptop.csv')
```

```python
df.head()
```

```python
df.drop(columns = ['Unnamed: 0','Unnamed: 0.1'],inplace = True)
df.columns
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Company'] = le.fit_transform(df['Company'])
df['TypeName'] = le.fit_transform(df['TypeName'])
df['Inches'] = le.fit_transform(df['Inches'])
df['ScreenResolution'] = le.fit_transform(df['ScreenResolution'])
df['Cpu'] = le.fit_transform(df['Cpu'].values)
df['Gpu'] = le.fit_transform(df['Gpu'].values)
df['OpSys'] = le.fit_transform(df['OpSys'].values)
df['Weight'] = le.fit_transform(df['Weight'].values)
```

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors = 5)
df['Company'] = imputer.fit_transform(df[['Company']])
df['TypeName'] = imputer.fit_transform(df[['TypeName']])
df['Inches'] = imputer.fit_transform(df[['Inches']])
df['ScreenResolution'] = imputer.fit_transform(df[['ScreenResolution']])
df['Cpu'] = imputer.fit_transform(df[['Cpu']])
df['Gpu'] = imputer.fit_transform(df[['Gpu']])
df['OpSys'] = imputer.fit_transform(df[['OpSys']])
df['Weight'] = imputer.fit_transform(df[['Weight']])
```

```python
df['Ram'] = df['Ram'].astype(str)
df['Memory'] = df['Memory'].astype(str)
```

```python
df['Ram'] = df['Ram'].str.extract('(\d+)').astype(float)
df['Memory'] = df['Memory'].str.extract('(\d+)').astype(float)
```

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors = 5)
df['Ram'] = imputer.fit_transform(df[['Ram']])
df['Memory'] = imputer.fit_transform(df[['Memory']])
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
Categorical_cols = ['Company','TypeName','Cpu','Gpu','OpSys','ScreenResolution']
Numerical_cols = ['Inches','Ram','Memory','Weight']
```

```python
x = df.drop(columns = ['Price'])
y = df['Price']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
x_test
```

```python
print(y_train.isnull().sum())
```

```python

print(y_train.isnull().sum())

y_train = y_train.dropna()
x_train = x_train.loc[y_train.index]


from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
y_train = imputer.fit_transform(y_train.values.reshape(-1, 1))
y_train = y_train.ravel()


lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python

print(np.isnan(y_test).sum())


non_nan_indices = ~np.isnan(y_test)
y_test = y_test[non_nan_indices]
x_test = x_test.loc[non_nan_indices]



y_pred = lr.predict(x_test)


mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f" mse:{mse}")
print(f" rmse:{rmse}")
```

```python
rf = RandomForestRegressor()
rf.fit(x_train,y_train)
y_pred = rf.predict(x_test)

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f" mse:{mse}")
print(f" rmse:{rmse}")
```
